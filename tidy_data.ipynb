# Example how to tidy data

```r
library(tidyverse)
```

```r
data <- read.csv('example_data.csv')
```

```r
data
```

```r
tidy_data <- data %>%
    pivot_longer(starts_with('grp'), # How to identify columns to pivot, e.g. starts with "grp"
                 names_to='sample',
                 values_to='intensity')
```

```r
tidy_data
```
